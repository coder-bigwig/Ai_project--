# 综合实验：自动驾驶视觉感知

本实验以图像预处理、边缘检测与可视化分析为主线，完成自动驾驶视觉感知基础流程实践。

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# TODO: 将 sample.png 替换为实验数据图像
# image = Image.open('sample.png').convert('L')
# gray = np.array(image)

# 暂用模拟图像数据演示流程
gray = np.random.randint(0, 255, (240, 320), dtype=np.uint8)
plt.figure(figsize=(6, 4))
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Input')
plt.axis('off')
plt.show()
```

```python
# 简单梯度边缘检测（不依赖 opencv）
gx = np.abs(np.diff(gray.astype(np.int16), axis=1, append=gray[:, -1:]))
gy = np.abs(np.diff(gray.astype(np.int16), axis=0, append=gray[-1:, :]))
edges = np.clip(gx + gy, 0, 255).astype(np.uint8)

plt.figure(figsize=(6, 4))
plt.imshow(edges, cmap='gray')
plt.title('Edge Map')
plt.axis('off')
plt.show()
```
