# 实验五：使用 Matplotlib 绘制可视化图表

本实验要求完成折线图、柱状图、散点图、子图布局与图例标注等基础可视化任务。

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)
y = x ** 2

plt.figure(figsize=(8, 4))
plt.plot(x, y, marker='o', label='y = x^2')
plt.title('Line Chart Demo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
